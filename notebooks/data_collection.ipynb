# Data Collection: S&P 500 Changes
This notebook scrapes historical S&P 500 constituent changes from Wikipedia.

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tables = pd.read_html(url)
changes_df = tables[1]
changes_df.to_csv('../data/sp500_changes.csv', index=False)
changes_df.head()
```
